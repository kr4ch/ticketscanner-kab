```python
import pandas as pd
import datetime

# Erwarte Ticket Informationen in einem Excel File, in einem Blatt (sheet)
#pfad_zu_ticket_excel_file = r"/home/tobias/Ticketscanner/Ticketverwaltung_Vorverkauf.xlsx"
pfad_zu_ticket_excel_file = r"/home/tobias/Downloads/Ticketverwaltung_Vorverkauf.xlsx"
sheet_name = 'Website'
output_excel_file = r"Ticketverwaltung_Vorverkauf_Scan.xlsx"
spalten = ['Ticket Nr.', 'Name', 'Scan Zeit']
zeile_mit_titeln = 7
zeile_header = zeile_mit_titeln-1
```

# Excel Datei einlesen

```python
data = pd.read_excel (pfad_zu_ticket_excel_file, sheet_name, sep='\s*,\s*', header=zeile_header)
#print (data)
tickets = pd.DataFrame(data, columns=['Ticket Nr.'])
#print (tickets)
```

```python
data.set_index("Ticket Nr.", inplace=True)
data.head()

exit_cond = True
print("Zum Stoppen 'q <ENTER>' eingeben.")
while exit_cond:
    text_input = input()
    if text_input == 'q':
        exit_cond = False
    else:
        if text_input in tickets.values:
            #print("Ticket in Liste vorhanden")
            #print(str(data.loc[text_input]))
            if not str(data.loc[text_input, "Scan Zeit"]) == 'nan':
                print("Dieses Ticket wurde bereits gescannt um " + str(data.loc[text_input, "Scan Zeit"]))
            else:
                print("Ticket ist korrekt und wurde erfasst.")
                data.loc[text_input, "Scan Zeit"] = datetime.datetime.now()
            #print(data.loc[text_input])
            data.head()
        else:
            print("Ticket UNBEKANNT!")

```

# Liste abspeichern

```python
#print(pfad_zu_output_excel_file[:-5]+"_"+str(datetime.datetime.now())+".xlsx")
speicherpfad = output_excel_Y = speicherpfad.replace(':', '_')
data.to_excel(speicherpfad)
```
